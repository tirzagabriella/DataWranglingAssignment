# Clean your text data using regex function

```python
import re

def clean_text(text):
    # Check if text is not NaN (float) before cleaning
    if pd.notna(text):
        # Remove extra whitespaces, leading, and trailing whitespaces
        cleaned_text = ' '.join(text.split())
        # Remove special characters and symbols using regex
        cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)
        return cleaned_text
    else:
        return text  # Return NaN as is

def clean_value(val):
    return re.sub(r"\.[^.]*$", "", str(val)) #clean unnecessary (value).0

df1['minutes_played'] = df1['minutes_played'].apply(clean_value) #clean .0 in minutes_played column
df1['team_id'] = df1['team_id'].apply(clean_value) #clean .0 in team_id column
df1['jersey_number'] = df1['jersey_number'].apply(clean_value) #clean .0 in jersey_number column
df1['game_id'] = df1['game_id'].apply(clean_value) #clean .0 in game_id column
df1
```

# Clean the text and nested json (before cleaning)

```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import re  # Import the regular expression module

# Function to safely parse JSON or handle malformed JSON
def safe_parse_json(json_str):
    try:
        return ast.literal_eval(json_str) if pd.notna(json_str) else None
    except (ValueError, SyntaxError):
        return None

# Load the CSV file from the URL into a DataFrame
url = 'https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv'

# Use the custom function to parse the problematic columns
df3 = pd.read_csv(url, converters={
    'belongs_to_collection': safe_parse_json,
    'genres': safe_parse_json,
    'production_companies': safe_parse_json,
    'production_countries': safe_parse_json,
    'spoken_languages': safe_parse_json
})

# Data Exploration
# print(df3.head())  # View the first few rows of the DataFrame
# print(df3.info())  # Get information about column data types and missing values
# print(df3.describe())  # Generate summary statistics

df3
```

# Clean the text and nested JSON (after cleaning)

```python
# Function to clean and standardize text data
def clean_text(text):
    # Check if text is not NaN (float) before cleaning
    if pd.notna(text):
        # Remove extra whitespaces, leading, and trailing whitespaces
        cleaned_text = ' '.join(text.split())
        # Remove special characters and symbols using regex
        cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)
        return cleaned_text
    else:
        return text  # Return NaN as is

def clean_value(val):
    return re.sub(r"\.[^.]*$", "", str(val)) #clean unnecessary (value).0

df3['runtime'] = df3['runtime'].apply(clean_value) #clean .0 in runtime column
df3['revenue'] = df3['revenue'].apply(clean_value) #clean .0 in revenue column
df3['vote_average'] = df3['vote_average'].apply(clean_value) #clean .0 in vote_average column
df3['vote_count'] = df3['vote_count'].apply(clean_value) #clean .0 in vote_count column

# Data Cleaning (Example: Removing Duplicates based on a subset of columns)
subset_columns = ['original_title', 'release_date', 'runtime', 'budget', 'revenue']
df3.drop_duplicates(subset=subset_columns, inplace=True)

# Clean and standardize text columns: 'overview' and 'title'
df3['overview'] = df3['overview'].apply(clean_text)
df3['title'] = df3['title'].apply(clean_text)

# Data Analysis (Example: Grouping by 'original_language' and calculating the average revenue)
def get_language(language_data):
    if language_data:
        return language_data[0]['iso_639_1']
    else:
        return None

df3['language'] = df3['spoken_languages'].apply(get_language)
average_revenue_by_language = df3.groupby('language')['revenue'].mean()

# Data Exploration
# print(df3.head())  # View the first few rows of the DataFrame
# print(df3.info())  # Get information about column data types and missing values
# print(df3.describe())  # Generate summary statistics

# Export Cleaned Data to a CSV file (optional)
cleaned_file_path = 'cleaned_movies_metadata.csv'
df3.to_csv(cleaned_file_path, index=False)

df3
```
